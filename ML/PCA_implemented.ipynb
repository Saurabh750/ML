```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
# instantiating
cancer = load_breast_cancer(as_frame=True)
# creating dataframe
df = cancer.frame
df.head(5)
```

```python
# Input features
X = df[cancer['feature_names']]
print('Inputs Dataframe shape   :', X.shape)

X_mean = X.mean()
X_std = X.std()
Z = (X-X_mean)/X_std

Z.head()
```

```python
N = 569
cov_Z_12 = 0
col_1_mean = Z.mean()['mean radius']
col_2_mean = Z.mean()['mean texture']
for i in range(N):
    cov_Z_12 += (Z['mean radius'][i] - col_1_mean)*(Z['mean texture'][i] - col_2_mean)
print(cov_Z_12/(N-1))

# I have calculated covariance of first and second column, which is same as we see
# in the covariance matrix calculated
```

```python
Z_cov = Z.cov()
```

```python
print(Z_cov)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(Z_cov)
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

```python
print(eigenvectors[0])
# I have printed first row of the matrix.
# Eigen vectors are the columns of this matrix. Look below how the sequence changes when we
# sort it corresponding to eigen values
```

```python
index = eigenvalues.argsort()[::-1]
print(index)
# Index will help access given array in descending order
```

```python
eigenvalues = eigenvalues[index]
# This will sort eigenvalues in descending order
```

```python
eigenvectors = eigenvectors[:,index]
# Eigenvectors shuffled so that they take indexes corresponding to their eigen values
```

```python
print(eigenvectors[0])
# The sequence has changed. This confirms that Eigen vectors are the columns of the matrix
```

```python
# I want to consider top 2 eigen vectors i.e. 
# squeeze the 29D into 2D thus
n_components = 2
Z_PC = eigenvectors[:,:n_components]
# Z_PC will contain 2 columns corresponding to the 2 Eigen vectors who capture maximum variance
print(Z_PC)
```

```python
# Converting into Dataframe
Z_PC = pd.DataFrame(X_PC,index = cancer['feature_names'], columns = ['PC1','PC2'])
Z_PC.head()
```

```python
# Now we just take the projection of each vector in Z onto our principal component vectors
# This is just like change of axes
Z_pca = np.dot(Z, Z_PC)
Z_pca_df = pd.DataFrame(Z_pca, columns = ['PC1', 'PC2'])
print(Z_pca_df)
```
